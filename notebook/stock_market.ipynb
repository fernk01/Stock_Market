Creamos un nuevo data frame con los columnas de interes al igual que excel.
- Leamos un archivo excel donde estan la accion y el precion de apertura. Creamos un data frame con el pnl 
y porcentaje de ganancia.

```python
df = pd.DataFrame(columns=['name_ticker', 'date', 'size', 'price_open', 'price_actual', 'PNL', 'poceentaje_PNL'])
# agrefo una columna con el nombre del ticker
df['name_ticker'] = stock_list['stock']
df['date'] = stock_list['date']
df['size'] = stock_list['size']
df['price_open'] = stock_list['price open']

# Obtenemos los datos en tiempo real.
data = pdr.get_data_yahoo(stock_list.stock.to_list(), period="1m",  interval='1m')

# Adignamos el precio actual a cada ticker.
data_dict = data['Close'].to_dict(orient='records')[0]
df['price_actual'] = df['name_ticker'].map(data_dict )

# Calculamos el PNL y el porcentaje de PNL
df['poceentaje_PNL'] = - (df['price_open'] - df['price_actual']) / df['price_open'] * 100
df['PNL'] = df['size'] * df['poceentaje_PNL'] / 100

df

```
